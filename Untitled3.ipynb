# Introdução a GANs

Os GANs aprendem uma distribuição de probabilidade de um conjunto de dados colocando duas redes neurais uma contra a outra.

Um modelo, o gerador, atua como um falsificador de imagens. Ele tenta criar imagens muito parecidas com o conjunto de dados. O outro modelo, o discriminador, age como a polícia e tenta detectar se as imagens geradas eram falsas ou não.

O que basicamente acontece é que o falsificador fica cada vez melhor em falsificações, enquanto a polícia fica cada vez melhor em detectar falsificações.

Efetivamente, esses dois modelos continuam tentando se superar, até que após muitas iterações, o gerador cria imagens indistinguíveis do conjunto de dados real.
O treinamento de GENss envolve dois objetivos:


* O discriminador maximiza a probabilidade de atribuir o rótulo correto aos exemplos de treinamento e às imagens geradas pelo gerador. Ou seja, o policial se torna melhor na diferenciação entre falsificações e pinturas reais.

* O gerador minimiza a probabilidade de o discriminador prever que o que ele gera é falso. Ou seja, o gerador se torna melhor na criação de falsificações

```
from keras.datasets import mnist
from keras.models import Sequential
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = (x_train.astype(np.float32) - 127.5)/127.5
x_train = x_train.reshape(60000, 784)
```

O gerador também precisa de vetores de entrada aleatórios para gerar imagens e, para isso, usaremos numpy

```
np.random.seed(10)

random_dim = 100
```

O gerador é apenas um modelo de rede neural de basico que pega um vetor de entrada aleatório e gera um vetor de 784 dim, que, quando remodelado (reshape), torna-se uma imagem de 28 * 28 pixels.

## Criando o Gerador

```
gerador = Sequential()
gerador.add(Dense(256, input_dim=random_dim, kernel_initializer='random_uniform'))
gerador.add(LeakyReLU(0.2))

gerador.add(Dense(512))
gerador.add(LeakyReLU(0.2))

gerador.add(Dense(1024))
gerador.add(LeakyReLU(0.2))

gerador.add(Dense(784, activation='tanh'))
gerador.compile(loss='binary_crossentropy', optimizer='adam')
```

O discriminador é outra rede neural que obtém a saída da rede anterior, um vetor de 784 dimensões e gera uma probabilidade entre 0 e 1 de que veio do conjunto de dados de treinamento

## Criando o Discriminador

```
discriminador = Sequential()
discriminador.add(Dense(1024, input_dim=784, kernel_initializer='random_uniform'))
discriminador.add(LeakyReLU(0.2))
discriminador.add(Dropout(0.3))

discriminador.add(Dense(512))
discriminador.add(LeakyReLU(0.2))
discriminador.add(Dropout(0.3))

discriminador.add(Dense(256))
discriminador.add(LeakyReLU(0.2))
discriminador.add(Dropout(0.3))

discriminador.add(Dense(1, activation='sigmoid'))
discriminador.compile(loss='binary_crossentropy', optimizer='adam')
```

Agora compilamos os dois modelos em uma única rede adversária, configurando a entrada como um vetor 100-dimensional e a saída como a saída do discriminador.


```
discriminador.trainable = False
    
gan_input = Input(shape=(random_dim,))

x = gerador(gan_input)

gan_output = discriminador(x)
    
gan = Model(inputs=gan_input, outputs=gan_output)
gan.compile(loss='binary_crossentropy', optimizer='adam')
```

## Implementação do DCGAN
Esta é a parte em que definimos nosso DCGAN. Definiremos nossa função de gerador de ruído, arquitetura de gerador e arquitetura de discriminador.

O bloco de código a seguir é uma função auxiliar para criar um vetor de ruído de comprimento predefinido para um gerador. Ele irá gerar o ruído que queremos converter em uma imagem usando nossa arquitetura de gerador.

```
def gen_noise(batch_size, noise_shape):
    return np.random.normal(0, 1, size=(batch_size,)+noise_shape)
```

## Arquitetura do gerador

O gerador é a parte mais crucial do GAN.
Aqui, criamos um gerador adicionando algumas camadas de convolução transpostas para aumentar a amostra do vetor de ruído em uma imagem.


Eu também uso muitas camadas Batchnorm com um momento de 0,5 e ativação LeakyReLU. Eu uso o otimizador Adam com β = 0,5.

```
def get_gen_normal(noise_shape):

    kernel_init = 'glorot_uniform'    
    gen_input = Input(shape = noise_shape) 
    
    # Transpose 2D conv layer 1. 
    generator = Conv2DTranspose(filters = 512, kernel_size = (4,4), strides = (1,1), 
                                padding = "valid", data_format = "channels_last", 
                                kernel_initializer = kernel_init)(gen_input)
    
    generator = BatchNormalization(momentum = 0.5)(generator)
    generator = LeakyReLU(0.2)(generator)
    
    # Transpose 2D conv layer 2.
    generator = Conv2DTranspose(filters = 256, kernel_size = (4,4), strides = (2,2), 
                                padding = "same", data_format = "channels_last", 
                                kernel_initializer = kernel_init)(generator)
    
    generator = BatchNormalization(momentum = 0.5)(generator)
    generator = LeakyReLU(0.2)(generator)
    
    # Transpose 2D conv layer 3.
    generator = Conv2DTranspose(filters = 128, kernel_size = (4,4), strides = (2,2), 
                                padding = "same", data_format = "channels_last", 
                                kernel_initializer = kernel_init)(generator)
    
    generator = BatchNormalization(momentum = 0.5)(generator)
    generator = LeakyReLU(0.2)(generator)
    
    # Transpose 2D conv layer 4.
    generator = Conv2DTranspose(filters = 64, kernel_size = (4,4), strides = (2,2), 
                                padding = "same", data_format = "channels_last", 
                                kernel_initializer = kernel_init)(generator)
    
    generator = BatchNormalization(momentum = 0.5)(generator)
    generator = LeakyReLU(0.2)(generator)
    
    # conv 2D layer 1.
    generator = Conv2D(filters = 64, kernel_size = (3,3), strides = (1,1), 
                       padding = "same", data_format = "channels_last", 
                       kernel_initializer = kernel_init)(generator)
    
    generator = BatchNormalization(momentum = 0.5)(generator)
    generator = LeakyReLU(0.2)(generator)
    

    generator = Conv2DTranspose(filters = 3, kernel_size = (4,4), strides = (2,2),
                                padding = "same", data_format = "channels_last", 
                                kernel_initializer = kernel_init)(generator)
    


    generator = Activation('tanh')(generator)
    

    gen_opt = Adam(lr=0.00015, beta_1=0.5)
    generator_model = Model(input = gen_input, output = generator)
    generator_model.compile(loss='binary_crossentropy', optimizer=gen_opt, metrics=['accuracy'])
    generator_model.summary()
    return generator_model
```

## Arquitetura discriminadora
Aqui está a arquitetura discriminadora em que uso uma série de camadas convolucionais e uma camada densa no final para prever se uma imagem é falsa ou não.

```
def get_disc_normal(image_shape=(64,64,3)):
    dropout_prob = 0.4
    kernel_init = 'glorot_uniform'
    dis_input = Input(shape = image_shape)
    
    # Conv layer 1:
    discriminator = Conv2D(filters = 64, kernel_size = (4,4), strides = (2,2), padding = "same", data_format = "channels_last", kernel_initializer = kernel_init)(dis_input)
    discriminator = LeakyReLU(0.2)(discriminator)
    # Conv layer 2:
    discriminator = Conv2D(filters = 128, kernel_size = (4,4), strides = (2,2), padding = "same", data_format = "channels_last", kernel_initializer = kernel_init)(discriminator)
    discriminator = BatchNormalization(momentum = 0.5)(discriminator)
    discriminator = LeakyReLU(0.2)(discriminator)
    # Conv layer 3:   
    discriminator = Conv2D(filters = 256, kernel_size = (4,4), strides = (2,2), padding = "same", data_format = "channels_last", kernel_initializer = kernel_init)(discriminator)
    discriminator = BatchNormalization(momentum = 0.5)(discriminator)
    discriminator = LeakyReLU(0.2)(discriminator)
    # Conv layer 4:
    discriminator = Conv2D(filters = 512, kernel_size = (4,4), strides = (2,2), padding = "same", data_format = "channels_last", kernel_initializer = kernel_init)(discriminator)
    discriminator = BatchNormalization(momentum = 0.5)(discriminator)
    discriminator = LeakyReLU(0.2)(discriminator)
    # Flatten
    discriminator = Flatten()(discriminator)
    # Dense Layer
    discriminator = Dense(1)(discriminator)
    # Sigmoid Activation
    discriminator = Activation('sigmoid')(discriminator)
    # Optimizer and Compiling model
    dis_opt = Adam(lr=0.0002, beta_1=0.5)
    discriminator_model = Model(input = dis_input, output = discriminator)
    discriminator_model.compile(loss='binary_crossentropy', optimizer=dis_opt, metrics=['accuracy'])
    discriminator_model.summary()
    return discriminator_model
```
